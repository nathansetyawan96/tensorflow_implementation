```python
import numpy as np
import gan
import matplotlib.pyplot as plt

from keras.datasets import mnist
from keras.utils.vis_utils import plot_model
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
for i in range(25):
    plt.subplot(5, 5, 1+i)
    plt.axis("off")
    plt.imshow(X_train[i], cmap="gray_r")
```

```python
discriminator = gan.discriminator()
discriminator.summary()
```

```python
latent_dim = 100
generator = gan.generator(latent_dim)
generator.summary()
```

```python
plot_model(discriminator, to_file = "discriminator.png", show_shapes=True, show_layer_names=True)
plot_model(generator, to_file = "generator.png", show_shapes=True, show_layer_names=True)
```
